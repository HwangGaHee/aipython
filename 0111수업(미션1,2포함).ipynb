그래프 여러개 나타내기-subplots()

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,4,9,16,25]

#fig는 객체, axs는 각 플랏의 위치 표시(?), 객체받는(?)
fig,axs=plt.subplots(2,2)

axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)
fig.suptitle('Drawing Graph')
#plt.title('sd') <- 이렇게 지정하면 제일 마지막으로 그린 그래프의 제목이 됨
plt.show()
```
